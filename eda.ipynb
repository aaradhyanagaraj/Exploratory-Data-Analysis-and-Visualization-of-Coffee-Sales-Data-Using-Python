"Exploratory Data Analysis and Visualization of Coffee Sales Data Using Python"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')

# Loading and reading dataset
df = pd.read_csv("index.csv")
print(df.head())

# Shape of the data
print(df.shape)

# Data information
print(df.info())

# Describing the data
print(df.describe())

# Column to list
print(df.columns.tolist())

# Check for missing values
print(df.isnull().sum())

# Checking duplicate values
print(df.duplicated().sum())

# Handling the missing values
df['card'] = df['card'].fillna(0)
print(df['card'])

# Checking data types
print(df.dtypes)

# Print summary statistics for numerical columns in the DataFrame
print(df.describe())

# Create a bar plot of coffee type frequencies
plt.figure(figsize=(10, 6))
df['coffee_name'].value_counts().plot(kind='bar')
plt.xlabel('Name')
plt.ylabel('Frequency')
plt.title('Frequency of Coffee Types')
plt.xticks(rotation=45)
plt.show()

# Create a histogram of the 'money' column
plt.figure(figsize=(10, 6))
df['money'].hist(bins=20)
plt.xlabel('Money')
plt.ylabel('Frequency')
plt.title('Distribution of Money')
plt.show()

# Create a pair plot of the DataFrame colored by 'cash_type'
sns.pairplot(df, hue='cash_type')
plt.show()

# Create a pie chart of total sales by coffee type
df.groupby('coffee_name')['money'].sum().plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%')
plt.ylabel('')  
plt.title('Total Sales by Coffee Type')
plt.show()

# Create a pie chart of total sales by payment method
df.groupby('cash_type')['money'].sum().plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%')
plt.ylabel('')  
plt.title('Total Sales by Payment Method')
plt.show()

# Create a scatter plot of total sales by coffee type
coffee_sales = df.groupby('coffee_name')['money'].sum().reset_index()
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='coffee_name', y='money', data=coffee_sales, palette='viridis')
plt.xlabel('Coffee Name')
plt.ylabel('Total Sales')
plt.title('Total Sales by Coffee Type')
plt.xticks(rotation=45)  
plt.show()

# Create a line plot of total sales by payment method
cash_sales = df.groupby('cash_type')['money'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='cash_type', y='money', data=cash_sales, palette='viridis')
plt.xlabel('Cash Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Payment Method')
plt.xticks(rotation=45)  
plt.show()
```
